```python
altindie_releases = []

for i in all_releases:
    i.fetch('year')
    i.fetch('id')
    i.fetch('title')
    i.fetch('label')
    i.fetch('format')
    i.fetch('genre')
    i.fetch('style')
    i.fetch('community')['want']
    i.fetch('community')['have']
    altindie_releases.append(i)
```

```python
years = []
ids = []
titles = []
labels = []
formats = []
genres = []
styles = []
community_data = []

for i in altindie_releases:

    year = i.fetch('year')
    years.append(year)

    id = i.fetch('id')
    ids.append(id)

    title = i.fetch('title')
    titles.append(title)

    label = i.fetch('label')
    labels.append(label)

    format = i.fetch('format')
    formats.append(format)

    genre = i.fetch('genre')
    genres.append(genre)

    style = i.fetch('style')
    styles.append(style)

    community = i.fetch('community')
    community_data.append(community)
```

```python
import pandas as pd

df1 = pd.DataFrame({
    'year': years,
    'id': ids,
    'artist and album name': titles,
    'release format': formats,
    'label': labels,
    'genre': genres,
    'subgenre': styles,
    'wants and haves': community_data
})
```
