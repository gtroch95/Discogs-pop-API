```python
major_cleaned = pd.read_csv('Major_releases_cleaned.csv')
```

```python
independent_cleaned = pd.read_csv('Independent_releases_cleaned.csv')
```

```python
self_cleaned = pd.read_csv('Self_released_cleaned.csv')
```

```python
# after reformatting the "label" columns for each CSV in OpenRefine, brought them back in to
# merge together

dataframes = [major_cleaned, independent_cleaned, self_cleaned]

labels = pd.concat(dataframes)
```

```python
vinyl = (labels[labels["release format"] == 'Vinyl'])
```

```python
cassette = (labels[labels["release format"] == 'Cassette'])
```

```python
cd = (labels[labels["release format"] == 'CD'])
```

```python
dataframes2 = [cd, vinyl, cassette]
```

```python
formats_labels_cleaned = pd.concat(dataframes2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.countplot(x='year', hue='year', legend=False, data=formats_labels_cleaned, palette="tab10")
plt.xticks(rotation=70)
plt.title("Total Pop Music Physical Releases")
```

```python
label = ("Major Labels", "Independent Labels", "Self-Released")
wants_haves = {
    'Want': (121654, 89619, 5456),
    'Have': (657598, 342462, 15648),
}

x = np.arange(len(label))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in wants_haves.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total Number of Wants/Haves')
ax.set_title('2025 Pop Releases - Wants vs. Haves by Label Type')
ax.set_xticks(x + width, label)
ax.legend(loc='upper right', ncols=2)
ax.set_ylim(0, 700000)

plt.show()
```
