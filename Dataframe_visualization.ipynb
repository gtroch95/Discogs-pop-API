```python
label = ("Major Labels", "Independent Labels", "Self-Released")
wants_haves = {
    'Want': (126405, 84868, 5456),
    'Have': (679802, 320258, 15648),
}

x = np.arange(len(label))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in wants_haves.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total Number of Wants/Haves')
ax.set_title('2025 Pop Release Wants vs. Haves by Label Type')
ax.set_xticks(x + width, label)
ax.legend(loc='upper right', ncols=2)
ax.set_ylim(0, 750000)

plt.show()
```

```python
release_format_graph = sns.countplot(x='year', hue='release format', data=formats_labels_cleaned, palette="tab10")
plt.xticks(rotation=70)
release_format_graph.legend(fontsize=10)
plt.title("Total Pop Music Physical Releases by Format")
plt.ylabel('Number of Releases')
```

```python
major_formats = (formats_labels_cleaned[formats_labels_cleaned["label"] == 'Major'])
indie_formats = (formats_labels_cleaned[formats_labels_cleaned["label"] == 'Independent'])
self_formats = (formats_labels_cleaned[formats_labels_cleaned["label"] == 'Not on Label'])
```

```python
major_release_graph = sns.countplot(x='year', hue='release format', data=major_formats, palette="tab10")
plt.xticks(rotation=70)
major_release_graph.legend(fontsize=10)
plt.title("Total Pop Music Physical Releases by Major Labels")
```

```python
indie_release_graph = sns.countplot(x='year', hue='release format', data=indie_formats, palette="tab10")
plt.xticks(rotation=70)
indie_release_graph.legend(fontsize=10)
plt.title("Total Pop Music Physical Releases by Independent Labels")
plt.ylabel('Number of Releases')
```

```python
self_release_graph = sns.countplot(x='year', hue='release format', data=self_formats, palette="tab10")
plt.xticks(rotation=70)
self_release_graph.set_ylim(0, 550)
self_release_graph.legend(fontsize=10)
plt.title("Total Pop Music Physical Releases by Artists Not on Label")
plt.ylabel('Number of Releases')
```

```python
releases_labeltype = sns.countplot(x='year', hue='label', data=formats_labels_cleaned, palette="tab10")
plt.xticks(rotation=70)
plt.title("Total Pop Music Physical Releases by Label Type")
releases_labeltype.legend(fontsize=7)
releases_labeltype.set_ylim(0, 8000)
plt.ylabel('Number of Releases')
sns.move_legend(releases_labeltype, "upper right")
```
